```python
import pandas as pd

# Carregar os dados do arquivo Excel fornecido pelo usuário
file_path = 'responses (1).xlsx'
data = pd.read_excel(file_path)

# Exibir as primeiras linhas para ter uma visão geral dos dados
data.head()
```

```python
# Limpeza e Preparação dos Dados

# Função para processar colunas de múltipla escolha:
# Vamos transformar as colunas de eventos e clubes de 'NaN' para 0 e o nome do evento/clube para 1.
def process_multichoice_columns(df, columns):
    for column in columns:
        df[column] = df[column].apply(lambda x: 0 if pd.isna(x) else 1)
    return df

# Identificar colunas que parecem ser de múltipla escolha baseadas no carregamento inicial.
multichoice_columns = [
    'ETHKampus Hacker House/space (ITU)', 'VM Day', 'Cartesi Workshop with YTU (12 Nov)',
    'Cartesi Online Workshop (7 Nov)', 'L2 Infra Day', 'L2 Days', 'Restaking Summit',
    'Other blockchain club events', 'I did not attend any side events', 'Other',
    'ITU', 'ODTU', 'BUchain', 'YTU', 'HER DAO Istanbul', 'I am not a part of any blockchain club', 'Other.1',
    'Prize money', 'Technology of the project', 'Alignment with my skills and expertise',
    'Personal interest and passion', 'Engagement and support from the project team',
    'Familiarity with the project', 'Other.2'
]

# Processar as colunas de múltipla escolha
data_processed = process_multichoice_columns(data, multichoice_columns)

# Exibir as primeiras linhas para verificar as mudanças
data_processed.head()
```

```python
# Análise Descritiva

# Contagem de frequência para o conhecimento de blockchain dos desenvolvedores
knowledge_counts = data_processed['How would you rate your current Web3/Blockchain knowledge as a developer?'].value_counts()

# Contagem de participantes para cada evento
events_columns = [
    'ETHKampus Hacker House/space (ITU)', 'VM Day', 'Cartesi Workshop with YTU (12 Nov)',
    'Cartesi Online Workshop (7 Nov)', 'L2 Infra Day', 'L2 Days', 'Restaking Summit',
    'Other blockchain club events', 'I did not attend any side events'
]
events_participation = data_processed[events_columns].sum()

# Contagem de participantes para cada clube de blockchain
clubs_columns = [
    'ITU', 'ODTU', 'BUchain', 'YTU', 'HER DAO Istanbul', 'I am not a part of any blockchain club', 'Other.1'
]
clubs_participation = data_processed[clubs_columns].sum()

# Contagem de fatores considerados importantes ao escolher uma track para hackear
factors_columns = [
    'Prize money', 'Technology of the project', 'Alignment with my skills and expertise',
    'Personal interest and passion', 'Engagement and support from the project team',
    'Familiarity with the project', 'Other.2'
]
factors_importance = data_processed[factors_columns].sum()

# Resumo das análises descritivas
knowledge_counts, events_participation, clubs_participation, factors_importance
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np

# Função para a análise de tópicos de texto livre
def topic_analysis(texts, n_topics=5, n_words=10):
    # Vetorizar o texto
    vec = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
    dtm = vec.fit_transform(texts.dropna())  # Descartar valores NaN

    # Instanciar e ajustar LDA
    lda = LatentDirichletAllocation(n_components=n_topics, random_state=0)
    lda.fit(dtm)

    # Obter os tópicos
    topics = dict()
    for i, topic in enumerate(lda.components_):
        top_features_ind = topic.argsort()[:-n_words - 1:-1]
        top_features = [vec.get_feature_names_out()[i] for i in top_features_ind]
        topics[i] = top_features

    return topics

# Análise de tópicos para a pergunta "What first comes to mind when thinking about Cartesi?"
cartesi_mind_topics = topic_analysis(data_processed['What first comes to mind when thinking about Cartesi?'])

# Análise de tópicos para a pergunta "In a few words, what did you like about it?"
liked_about_it_topics = topic_analysis(data_processed['In a few words, what did you like about it?'])

cartesi_mind_topics, liked_about_it_topics
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurando o estilo dos gráficos
sns.set(style="whitegrid")

# Função para plotar gráficos de barras
def plot_bar(data, title, xlabel, ylabel):
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=data.values, y=data.index)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.show()

# Plotar o conhecimento de Blockchain dos desenvolvedores
plot_bar(knowledge_counts, 'Developer Blockchain Knowledge', 'Count', 'Knowledge Level')

# Plotar a participação nos eventos
plot_bar(events_participation, 'Participation in Side Events', 'Count', 'Events')

# Plotar a participação nos clubes de blockchain
plot_bar(clubs_participation, 'Participation in Blockchain Clubs', 'Count', 'Blockchain Clubs')

# Plotar os fatores considerados ao escolher uma track para hackear
plot_bar(factors_importance, 'Factors Considered for Hacking Track Choice', 'Count', 'Factors')
```

```python
# Para visualizar os dados de respostas abertas, podemos usar um Word Cloud.
# Isso nos ajudará a ver as palavras mais comuns em cada conjunto de respostas.

from wordcloud import WordCloud

# Função para plotar Word Cloud
def plot_word_cloud(topics, title):
    wc = WordCloud(width=800, height=400, background_color='white', max_words=200)
    # Combine text in the topics
    text = " ".join([" ".join(words) for words in topics.values()])
    wc.generate(text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation='bilinear')
    plt.title(title, fontsize=18)
    plt.axis('off')
    plt.show()

# Plotar Word Cloud para a pergunta "What first comes to mind when thinking about Cartesi?"
plot_word_cloud(cartesi_mind_topics, 'Word Cloud for Thoughts on Cartesi')

# Plotar Word Cloud para a pergunta "In a few words, what did you like about it?"
plot_word_cloud(liked_about_it_topics, 'Word Cloud for Likes About the Game Demo')
```

```python
from scipy.stats import chi2_contingency

# Função para testar a correlação entre duas variáveis categóricas usando o teste Chi-quadrado
def test_chi2_independence(df, col1, col2, verbose=True):
    contingency_table = pd.crosstab(df[col1], df[col2])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    if verbose:
        print(f"Chi-Square Test between {col1} and {col2}")
        print(f"Chi-Square statistic: {chi2}")
        print(f"p-value: {p}\n")
    return chi2, p

# Convertendo o nível de conhecimento de blockchain em uma variável numérica para correlação
knowledge_mapping = {
    "Novice (Just starting)": 1,
    "Beginner (Basic understanding, limited practical exposure)": 2,
    "Intermediate (Solid grasp, some project/ task experience)": 3,
    "Advanced (Strong knowledge, multiple project experience)": 4,
    "Expert (Extensive, in-depth knowledge, thought leader)": 5
}
data_processed['Knowledge_Numeric'] = data_processed['How would you rate your current Web3/Blockchain knowledge as a developer?'].map(knowledge_mapping)

# Teste de correlação entre nível de conhecimento e participação em eventos
knowledge_event_correlation = data_processed[events_columns + ['Knowledge_Numeric']].corr()['Knowledge_Numeric'][:-1]

# Teste de correlação entre nível de conhecimento e fatores de decisão de trilha
knowledge_factor_correlation = data_processed[factors_columns + ['Knowledge_Numeric']].corr()['Knowledge_Numeric'][:-1]

# Teste de correlação entre nível de conhecimento e participação em clubes de blockchain
knowledge_club_correlation = data_processed[clubs_columns + ['Knowledge_Numeric']].corr()['Knowledge_Numeric'][:-1]

# Teste de Chi-quadrado entre participação em clubes de blockchain e eventos específicos
chi2_clubs_events = {}
for club in clubs_columns:
    for event in events_columns:
        chi2, p = test_chi2_independence(data_processed, club, event, verbose=False)
        chi2_clubs_events[(club, event)] = p

# Correlações mais fortes
knowledge_event_correlation_abs = knowledge_event_correlation.abs().sort_values(ascending=False)
knowledge_factor_correlation_abs = knowledge_factor_correlation.abs().sort_values(ascending=False)
knowledge_club_correlation_abs = knowledge_club_correlation.abs().sort_values(ascending=False)

knowledge_event_correlation_abs, knowledge_factor_correlation_abs, knowledge_club_correlation_abs, chi2_clubs_events
```

```python
# Vamos criar visualizações para as correlações e associações encontradas.

# Função para plotar gráficos de correlação
def plot_correlation(data, title, xlabel):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=data.values, y=data.index, palette="vlag")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Variables')
    plt.show()

# Plotar a correlação entre o nível de conhecimento e a participação em eventos
plot_correlation(knowledge_event_correlation_abs, 'Correlation with Blockchain Knowledge: Event Participation', 'Correlation Coefficient')

# Plotar a correlação entre o nível de conhecimento e os fatores de decisão de trilha
plot_correlation(knowledge_factor_correlation_abs, 'Correlation with Blockchain Knowledge: Hackathon Track Choice Factors', 'Correlation Coefficient')

# Plotar a correlação entre o nível de conhecimento e a participação em clubes de blockchain
plot_correlation(knowledge_club_correlation_abs, 'Correlation with Blockchain Knowledge: Blockchain Club Participation', 'Correlation Coefficient')
```
